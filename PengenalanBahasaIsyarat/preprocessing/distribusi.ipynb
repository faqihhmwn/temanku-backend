# Distribusi Dataset

```python
import os
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
BASE_DIR = "../dataset_fiks/fiksAug"   # sesuaikan root dataset
SPLITS = ["train", "val", "test"]
IMAGE_EXT = [".jpg", ".jpeg", ".png"]
```

## img & preview

```python
def count_images(split):
    img_base = os.path.join(BASE_DIR, split, "img")
    counts = {}

    for cls in os.listdir(img_base):
        cls_path = os.path.join(img_base, cls)
        if not os.path.isdir(cls_path):
            continue

        counts[cls] = len([
            f for f in os.listdir(cls_path)
            if os.path.splitext(f)[1].lower() in IMAGE_EXT
        ])

    return counts

def count_preview(split):
    img_base = os.path.join(BASE_DIR, split, "preview")
    counts = {}

    for cls in os.listdir(img_base):
        cls_path = os.path.join(img_base, cls)
        if not os.path.isdir(cls_path):
            continue

        counts[cls] = len([
            f for f in os.listdir(cls_path)
            if os.path.splitext(f)[1].lower() in IMAGE_EXT
        ])

    return counts
```

```python
# img
for split in SPLITS:
    print(f"\n=== {split.upper()} ===")
    counts = count_images(split)
    total = 0

    for cls, num in sorted(counts.items()):
        print(f"Kelas {cls}: {num}")
        total += num

    print(f"Total {split}: {total}")
```

```python
# preview
for split in SPLITS:
    print(f"\n=== {split.upper()} ===")
    counts = count_images(split)
    total = 0

    for cls, num in sorted(counts.items()):
        print(f"Kelas {cls}: {num}")
        total += num

    print(f"Total {split}: {total}")
```

```python
# img
rows = []

for split in SPLITS:
    counts = count_images(split)
    for cls, num in counts.items():
        rows.append({
            "Split": split,
            "Kelas": cls,
            "Jumlah": num
        })

df = pd.DataFrame(rows)
df[df["Split"] == "train"]

```

```python
# preview
rows = []

for split in SPLITS:
    counts = count_preview(split)
    for cls, num in counts.items():
        rows.append({
            "Split": split,
            "Kelas": cls,
            "Jumlah": num
        })

df_preview = pd.DataFrame(rows)
df_preview[df_preview["Split"] == "train"]

```

```python
import matplotlib.pyplot as plt
import numpy as np

kelas = df["Kelas"].unique()
splits = SPLITS  # ["train", "val", "test"]

x = np.arange(len(kelas))          # posisi kelas
width = 0.25                       # lebar bar

plt.figure(figsize=(10, 5))

for i, split in enumerate(splits):
    sub = df[df["Split"] == split]
    plt.bar(
        x + i * width,
        sub["Jumlah"],
        width,
        label=split
    )

plt.xlabel("Kelas")
plt.ylabel("Jumlah Data")
plt.title("Persebaran Dataset Train, Validation, dan Test per Kelas")
plt.xticks(x + width, kelas, rotation=45, ha="right")
plt.legend()
plt.tight_layout()
plt.show()
```
